````python
# @title
from IPython.display import display, Markdown

latex_equation = r"""
**Mathematical Justification (Variance Shrinks)**

As the number of trials \( n \) increases, the standard error decreases as:

$$
\text{Standard Error} = \frac{\sigma}{\sqrt{n}}
$$

**Where:**

\(\sigma\) is the standard deviation of a single trial.
\(n\) is the number of trials (or shots in a quantum circuit).

**For large \( n \), the variance in probability estimation follows:**

$$
\text{Error} \approx \frac{1}{\sqrt{n}}
$$

**Examples:**

$$
\text{For } n = 100, \quad \text{Error} \approx \frac{1}{\sqrt{100}} = 10\%
$$

$$
\text{For } n = 10,000, \quad \text{Error} \approx \frac{1}{\sqrt{10,000}} = 1\%
$$

$$
\text{For } n = 1,000,000, \quad \text{Error} \approx \frac{1}{\sqrt{1,000,000}} = 0.1\%
$$

Thus, as \( n \) increases, the variance in measured outcomes decreases, making the observed probability closer to the theoretical probability.
"""

display(Markdown(latex_equation))
````

### Here we will first see the optimisation of the qiskit then understand use of barrier in a circuit

````python
qbit = QuantumRegister(1,"Q")
ebit0 = QuantumRegister(1,"E0")
ebit1 = QuantumRegister(1,"E1")
A = ClassicalRegister(1,"A")
B = ClassicalRegister(1,"B")

circuit = QuantumCircuit(qbit,ebit1,ebit0,A,B)
circuit.cx(qbit,ebit1)
circuit.h(qbit)

circuit.measure(qbit,A)
circuit.measure(ebit1,B)

with circuit.if_test((A,1)):
  circuit.x(ebit0)

with circuit.if_test((B,1)):
  circuit.z(ebit0)

display(circuit.draw(output = 'mpl'))
````

````python
qbit = QuantumRegister(1,"Q")
ebit0 = QuantumRegister(1,"E0")
ebit1 = QuantumRegister(1,"E1")
A = ClassicalRegister(1,"A")
B = ClassicalRegister(1,"B")

circuit = QuantumCircuit(qbit,ebit1,ebit0,A,B)
circuit.cx(qbit,ebit1)
circuit.h(qbit)
circuit.barrier()

circuit.measure(qbit,A)
circuit.measure(ebit1,B)
circuit.barrier()

with circuit.if_test((A,1)):
  circuit.x(ebit0)

with circuit.if_test((B,1)):
  circuit.z(ebit0)

display(circuit.draw(output = 'mpl'))
````

The functions of `QuantumCircuit` in Qiskit allow users to construct, manipulate, and analyze quantum circuits. Here’s a categorized breakdown of these methods:

---

### **1. Circuit Construction & Modification**
These methods help in defining, modifying, and structuring the quantum circuit.

- **`add_bits`** – Adds classical or quantum bits to the circuit.
- **`add_register`** – Adds a quantum or classical register to the circuit.
- **`add_var`** – Declares a new classical variable for a circuit.
- **`add_uninitialized_var`** – Declares an uninitialized variable.
- **`append`** – Appends a gate or instruction to the circuit.
- **`compose`** – Merges another circuit or instruction into this circuit.
- **`copy`** – Creates a copy of the circuit.
- **`copy_empty_like`** – Creates an empty copy with the same registers.
- **`initialize`** – Sets an initial quantum state.
- **`inverse`** – Returns the inverse of the circuit.
- **`power`** – Raises a circuit to a given power.
- **`tensor`** – Tensor product (Kronecker product) with another circuit.
- **`repeat`** – Repeats the circuit multiple times.
- **`reverse_ops`** – Reverses the order of operations in the circuit.
- **`reverse_bits`** – Reverses the order of qubits.

---

### **2. Quantum Operations (Gates)**
These functions apply quantum gates to qubits.

- **Two-Qubit & Multi-Qubit Gates:**
  - `cx`, `cy`, `cz` – Controlled-X (CNOT), Y, and Z gates.

  - `cp`, `crx`, `cry`, `crz` – Controlled phase and rotation gates.
  - `cs`, `csdg` – Controlled-S and its inverse.
  - `cswap` – Controlled SWAP gate.
  - `csx` – Controlled sqrt(X) gate.
  - `cu`, `cu3` – Controlled U-gates.
  - `dcx` – Double CNOT gate.

  - `ecr` – Echoed cross-resonance gate.
  - `mcp`, `mcrx`, `mcry`, `mcrz`, `mcx` – Multi-controlled gates.
  - `rccx`, `rcccx` – Recursive C3X and C4X gates.
  - `rxx`, `ryy`, `rzx`, `rzz` – Two-qubit rotation gates.
  - `swap` – SWAP gate.

These methods measure qubits and apply classical logic.

### **Detailed Backend Working of ************`qasm_simulator`************ (Qiskit) with Gate Combinations and SWAPs**

The `qasm_simulator` in **Qiskit’s Aer module** is a high-performance simulator for quantum circuits. It processes circuits step by step, from ingestion to execution, including **compilation, optimization, and execution**.

---

## **1. Circuit Design & Ingestion**

When you create a circuit using `QuantumCircuit`, it exists as an **abstract representation** of quantum operations.

Example:

```python
from qiskit import QuantumCircuit, Aer, execute

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()
```

- **`qc.h(0)`** → Applies a Hadamard gate to qubit 0.
- **`qc.cx(0, 1)`** → Applies a CNOT gate with qubit 0 as control and qubit 1 as target.
- **`qc.measure_all()`** → Measures both qubits.

At this stage, Qiskit **stores this circuit in a data structure** but hasn't compiled or optimized it yet.

---

## **2. Backend Selection**

Now, we select a simulator backend:

```python
backend = Aer.get_backend('qasm_simulator')
```

💡 **Backend Selection Insights:**

- The **backend** is responsible for:
  - **Transpilation (if required)**
  - **Execution of quantum instructions**
  - **Simulating noise (if enabled)**

---

## **3. Execution Request (********`execute()`********\*\*\*\*)**

Now, we execute the circuit:

```python
job = execute(qc, backend, shots=1024)
result = job.result()
```

- **`shots=1024`** → Runs **1024 times**, aggregating results for probabilities.
- **Qiskit creates a "Job" object**:
  - Encapsulates the circuit and backend settings.
  - Handles **parallel execution**.

---

## **4. Preprocessing Before Execution**

### **4.1. Transpilation and Gate Combinations**

Before execution, Qiskit **transpiles** the circuit:

```python
from qiskit import transpile

transpiled_qc = transpile(qc, backend)
```

#### **What happens during transpilation?**

- Converts the circuit into **backend-compatible instructions**.
- If a **coupling map** is specified, it **inserts SWAP gates** to match connectivity.
- **Optimizes the circuit** by reducing gate count where possible.
- Converts **high-level gates into basic gates** (`h`, `cx`, `rz`, `sx`).

#### **Gate Combination Example**

Optimizations include merging adjacent single-qubit gates where possible.
For example:

```
X H H → X
```

---

### **4.2. SWAP Gates Due to Couplings**

If the device backend only allows **nearest-neighbor interactions**, additional SWAP gates may be inserted:

Before transpilation:

```
0: ───H───@───
           |
1: ───────X───
```

After transpilation (assuming restricted connectivity):

```
0: ───H───@───X───@───
           |       |
1: ───────X───────X───
```

🚀 **Why does Qiskit insert SWAP gates?**

- If two qubits **aren’t physically connected**, SWAP gates bring them closer **logically**.

---

## **5. Circuit Compilation**

After transpilation, the **backend compiles** the circuit into **low-level machine code**:

Example QASM output:

```
OPENQASM 2.0;
include "qelib1.inc";
qreg q[2];
creg c[2];
h q[0];
cx q[0], q[1];
measure q[0] -> c[0];
measure q[1] -> c[1];
```

---

## **6. Execution on the Simulator**

### **6.1. Statevector Evolution**

The simulator applies matrix multiplications:

1. **Hadamard Gate (********`H`********\*\*\*\*)**
   $$
   |0\rangle \rightarrow \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)
   $$
2. **CNOT Gate (********`CX`********\*\*\*\*)**
   $$
   \frac{1}{\sqrt{2}}(|00\rangle + |10\rangle) \rightarrow \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)
   $$

Final state:

$$
\frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)
$$

### **6.2. Measurement Simulation**

If **`shots=1024`**, the results will approximate:

```
{'00': 512, '11': 512}
```

---

## **7. Postprocessing**

### **7.1. Collecting Results**

```python
counts = result.get_counts()
print(counts)
```

Example output:

```python
{'00': 502, '11': 522}
```

### **7.2. Visualization**

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

---

## **8. Backend Execution Model**

💡 **Providers manage backends.**

```python
from qiskit import Aer, IBMQ

IBMQ.load_account()
provider = IBMQ.get_provider()
backend = provider.get_backend('ibmq_qasm_simulator')
```

- **`Aer`** → Local simulator.
- **`IBMQ`** → Cloud-based IBM Quantum systems.

### **Backend Compilation Process**

1. **Input Circuit** → (Gates & Measurements)
2. **Transpilation** → (Optimized for the device)
3. **Compilation** → (Converts to low-level QASM)
4. **Execution** → (Simulates state evolution)
5. **Measurement** → (Collapses to classical bits)

---

## **Summary**

| **Stage**             | **What Happens?**                                |
| --------------------- | ------------------------------------------------ |
| **Circuit Design**    | User defines a `QuantumCircuit`                  |
| **Backend Selection** | Choose `qasm_simulator`                          |
| **Execution Request** | `execute(qc, backend, shots=1024)`               |
| **Transpilation**     | Converts circuit into backend-compatible form    |
| **SWAP Insertion**    | Ensures connectivity in restricted architectures |
| **Compilation**       | Optimizes & converts to QASM                     |
| **Execution**         | Simulates quantum state evolution                |
| **Measurement**       | Samples classical results                        |
| **Postprocessing**    | Extracts and visualizes results                  |

---

### 🚀 **Key Takeaways**

✅ **Simulators follow a structured pipeline** from circuit ingestion to execution.\
✅ **Transpilation optimizes gates and ensures connectivity.**\
✅ **Gate combinations reduce redundant operations.**\

````python
import qiskit
import inspect

def list_qiskit_modules():
    """Lists all modules in the qiskit package and their functions."""
    public_modules = [attr for attr in dir(qiskit) if not attr.startswith("_")]

    for module_name in public_modules:
        try:
            module_obj = getattr(qiskit, module_name)
            functions = [attr for attr in dir(module_obj) if inspect.isfunction(getattr(module_obj, attr, None))]

            print(f"\n\033[1mModule: {module_name}\033[0m")  # Print module name in bold
            print("=" * 60)

            if functions:
                for func in functions:
                    print(f"- {func}")  # Print function name
            else:
                print("No functions found.")

        except Exception as e:
            print(f"Skipping {module_name} (not accessible): {e}")

# Run the function
list_qiskit_modules()
````

````python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from sympy import Matrix, nsimplify, latex  # For fractions & LaTeX
from IPython.display import display, Math  # For LaTeX rendering in Jupyter

# Circuit for QASM simulator (with measurements)
qc_qasm = QuantumCircuit(2, 2)
qc_qasm.h(0)
qc_qasm.cx(0, 1)
qc_qasm.measure([0, 1], [0, 1])  # Add measurement for QASM

# Circuit for Statevector & Unitary simulators (no measurements)
qc_no_meas = QuantumCircuit(2)
qc_no_meas.h(0)
qc_no_meas.cx(0, 1)

# List of simulators
backends = [
    "qasm_simulator",
    "statevector_simulator",
    "unitary_simulator",
]

# Run the circuit on each simulator
for backend_name in backends:
    backend = Aer.get_backend(backend_name)

    if backend_name == "qasm_simulator":
        transpiled_qc = transpile(qc_qasm, backend)  # Use circuit with measurements
    else:
        transpiled_qc = transpile(qc_no_meas, backend)  # Use circuit without measurements

    # Execute the circuit
    job = backend.run(transpiled_qc)
    result = job.result()

    # Display results based on backend
    print(f"\nResults from {backend_name}:")

    if backend_name == "qasm_simulator":
        print("Measurement counts:", result.get_counts())

    elif backend_name == "statevector_simulator":
        statevector = result.get_statevector()
        fraction_statevector = [nsimplify(complex(x), tolerance=1e-10) for x in statevector]  # Convert to fractions
        print("Statevector (Fraction Form):")
        display(Math(latex(fraction_statevector)))  # LaTeX rendering

    elif backend_name == "unitary_simulator":
        unitary_matrix = result.get_unitary()
        fraction_unitary = Matrix(unitary_matrix).applyfunc(lambda x: nsimplify(complex(x), tolerance=1e-10))
        print("Unitary Matrix (Fraction Form):")
        display(Math(latex(fraction_unitary)))  # LaTeX rendering
````

````python
from qiskit import QuantumCircuit,  transpile
from qiskit.visualization import plot_histogram

def create_bell_state(state):
    qc = QuantumCircuit(2)

    # Apply the common Hadamard and CNOT gates
    qc.h(0)
    qc.cx(0, 1)

    # Modify circuit to get different Bell states
    if state == "Phi-":
        qc.z(0)  # Apply Z gate to first qubit
    elif state == "Psi+":
        qc.x(0)  # Apply X gate to first qubit
    elif state == "Psi-":
        qc.x(0)
        qc.z(0)  # Apply X and Z to first qubit

    return qc

# Generate and display circuits for all Bell states
bell_states = ["Phi+", "Phi-", "Psi+", "Psi-"]
for state in bell_states:
    qc = create_bell_state(state)
    print(f"Bell State |{state}⟩:")
    display(qc.draw("mpl"))  # Visualize the quantum circuit


````
